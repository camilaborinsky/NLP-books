```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_metadata = pd.read_csv('gutenberg.csv')
```

```python
# plot missing values sum per column
plt.figure(figsize=(5, 5))

sns.barplot(x=df_metadata.isnull().sum().index, y=df_metadata.isnull().sum().values)
#rotate labes 45Â°
plt.xticks(rotation=45)
```

```python
#plot value counts of language column
count = df_metadata['Language'].value_counts()
#group all languages with less than 1000 books in 'other'
count['other'] = count[count < 1000].sum()
count = count.drop(count[count < 1000].index)

plt.figure(figsize=(10, 10))
sns.barplot(x=count.index, y=count.values)
#rotate 90Â°
plt.xticks(rotation=90)
```

```python
#plot value counts of type
plt.figure(figsize=(10, 5))
sns.barplot(x=df_metadata['Type'].value_counts().index, y=df_metadata['Type'].value_counts().values)
```

```python
#filter type book
df_metadata = df_metadata[df_metadata['Type'] == 'Text']
#filter only english books
df_metadata = df_metadata[df_metadata['Language'] == 'en']
df_metadata = df_metadata[['Text#','Authors','Title']]

```

```python
#plot missing values sum per column
plt.figure(figsize=(5, 5))
sns.barplot(x=df_metadata.isnull().sum().index, y=df_metadata.isnull().sum().values)
```

```python
#drop missing values
df_metadata = df_metadata.dropna()
```
